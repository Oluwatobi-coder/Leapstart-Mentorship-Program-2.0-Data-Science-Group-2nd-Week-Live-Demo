### Adult Income Prediction Dataset

**Aim**: To apply explatory data analysis to uncover important insights from the dataset of individuals from the 1994 U.S. Census database.

**Source**: The *Adult Income Prediction Dataset* dataset was obtained from [Kaggle](https://www.kaggle.com/datasets/mosapabdelghany/adult-income-prediction-dataset?select=adult.csv) website

### 1. Import libraries

```python
# import required python libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Loading of Dataset

```python
# load the data
adult_df = pd.read_csv('../data/adult.csv')
```

```python
# histogram plot for distribution of age

# extract the age column of the dataset
age_col = adult_df['age']

# plot the histogram
plt.hist(x=age_col, bins=20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
# histogram for distribution of hours per week

# extract the hours worked per week column
hours_col = adult_df['hours.per.week']

# plot the histogram
plt.hist(x=hours_col, bins=20)
plt.title('Distribution of Work Hours')
plt.xlabel('Hours')
plt.ylabel('Frequency')

plt.show()
```

```python
# distribution of adults by education level

# extract the counts for each education level
edu_col = adult_df['education'].value_counts()

# specify the plot size 
plt.figure(figsize=(20, 10))

# plot the horizontal bar chart 
bar_chrt_edu = plt.barh(edu_col.index, edu_col.values)
plt.title('Distribution of Education Level')
plt.ylabel('Education Level')
plt.xlabel('Count')

# add labels on top of the bars
plt.bar_label(bar_chrt_edu, fmt='%.0f')

plt.show()
```

```python
# distribution of adults by work class

# extract the counts for each work class
wrkclass_col = adult_df['workclass'].value_counts()

# specify the plot size 
plt.figure(figsize=(20, 10))

# plot the horizontal bar chart 
bar_chrt_wrk = plt.barh(y=wrkclass_col.index, width=wrkclass_col.values)
plt.title('Distribution of Work Class')
plt.xlabel('Count')
plt.ylabel('Work Class')

# add labels on top of the bars
plt.bar_label(bar_chrt_wrk, fmt='%.0f')

plt.show()
```

```python
# distribution of adults by gender

# extract the counts for each gender
gender_col = adult_df['sex'].value_counts()

# plot the horizontal bar chart
plt.barh(y=gender_col.index, width=gender_col.values)
plt.title('Distribution of Gender')
plt.xlabel('Count')
plt.ylabel('Gender')

plt.show()
```

```python
# distribution of income across the adult population

# extract the counts for each salary range
inc_dist = adult_df['income'].value_counts()

# extract the income categories and corresponding counts
x, y = inc_dist.index, inc_dist.values

# plot the bar chart
plt.bar(x, y)
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Count')

plt.show()

```

```python
# distribution of adults by race

# extract the counts for each race
race_col = adult_df['race'].value_counts()

# plot the horizontal bar chart
plt.barh(y=race_col.index, width=race_col.values)
plt.title('Distribution of Race')
plt.xlabel('Count')
plt.ylabel('Race')

plt.show()
```

```python
# distribution of adults by native country

# extract the counts for each native country
crty_col = adult_df['native.country'].value_counts()

# increase the plot size
plt.figure(figsize=(10, 12))

# plot the horizontal bar chart
plt.barh(y=crty_col.index, width=crty_col.values)
plt.title('Distribution of Native Country')
plt.xlabel('Count')
plt.ylabel('Native Country')

plt.show()
```

### 4. Cleaning the Dataset

```python
# check for null values
adult_df.isna().sum()
```

```python
# remove the occurence of "?"" from the dataset

rows_to_maintain = adult_df.astype(str).apply(lambda x: x.str.contains(r'\?').any(), axis=1)

new_adult_income_df = adult_df[~rows_to_maintain]

# view the first few rows of the clean dataset
new_adult_income_df.head(10)
```

```python
# drop the current index of the clean dataset
new_adult_income_df.reset_index(inplace=True, drop=True)

# view the first few rows of the clean dataset
new_adult_income_df.head(10)
```

```python
# check the size of the clean dataset
new_adult_income_df.shape
```

```python
# check the size of the original dataset
adult_df.shape
```

```python
# determine the percentage of data left after cleaning

# no of records in original dataset
len_raw = len(adult_df)

# no of records in clean dataset
len_clean = len(new_adult_income_df)

# calculate the percentage of data left
percent_rem_data = (len_clean / len_raw) * 100

percent_rem_data
```
